```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('Iris.csv',index_col="Id")
data_set = dataset.loc[(dataset.Species=='Iris-setosa')]
data_ver = dataset.loc[(dataset.Species=='Iris-versicolor')]
data_vir = dataset.loc[(dataset.Species=='Iris-virginica')]
X = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:5].values
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(data_set['PetalLengthCm'],label='Iris-setosa',shade=True)
sns.kdeplot(data_ver['PetalLengthCm'],label='Iris-versicolor',shade=True)
sns.kdeplot(data_vir['PetalLengthCm'],label='Iris-virginica',shade=True)
plt.title('Distribution of Petal Lengths in Cm')
plt.show()

plt.figure(figsize=(3,3))
sns.kdeplot(data_set['PetalWidthCm'],label='Iris-setosa',shade=True)
sns.kdeplot(data_ver['PetalWidthCm'],label='Iris-versicolor',shade=True)
sns.kdeplot(data_vir['PetalWidthCm'],label='Iris-virginica',shade=True)
plt.title('Distribution of Petal Widths in Cm')
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(3,3))
sns.kdeplot(data_set['SepalLengthCm'],label='Iris-setosa',shade=True)
sns.kdeplot(data_ver['SepalLengthCm'],label='Iris-versicolor',shade=True)
sns.kdeplot(data_vir['SepalLengthCm'],label='Iris-virginica',shade=True)
plt.title('Distribution of Sepal Lengths in Cm')
plt.show()

plt.figure(figsize=(3,3))
sns.kdeplot(data_set['SepalWidthCm'],label='Iris-setosa',shade=True)
sns.kdeplot(data_ver['SepalWidthCm'],label='Iris-versicolor',shade=True)
sns.kdeplot(data_vir['SepalWidthCm'],label='Iris-virginica',shade=True)
plt.title('Distribution of Sepal Widths in Cm')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.67,random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion = 'entropy',random_state=0)
classifier.fit(X_train,y_train)
y_pred_random_forest = classifier.predict(X_test)
accuracy_score(y_test,y_pred_random_forest)
```

```python
from sklearn.svm import SVC
classifier = SVC(random_state=0)
classifier.fit(X_train,y_train)
y_pred_svm = classifier.predict(X_test)
accuracy_score(y_test,y_pred_svm)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
y_pred_logistic_regression = classifier.predict(X_test)
accuracy_score(y_test,y_pred_logistic_regression)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train,y_train)
y_pred_knn = classifier.predict(X_test)
accuracy_score(y_test,y_pred_knn)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
y_pred_naive_bayes = classifier.predict(X_test)
accuracy_score(y_test,y_pred_naive_bayes)
```
